```python
# a plot of a ReLU activation function and its gradient in the same graph
import numpy as np
from matplotlib import pyplot
from math import exp
# rectified linear function
def rectified(x):
    return max(0.0, x)
# derivative of rectified linear function
def rectified_derivative(x):
    return 0.0 if x < 0.0 else 1.0

# Define the x range
x_range = np.linspace(-5, 5, 1000)

# line plot of rectified linear function
pyplot.plot(x_range, [rectified(x) for x in x_range], label='ReLU')
# line plot of derivative of rectified linear function
pyplot.plot(x_range, [rectified_derivative(x) for x in x_range], label='Gradient of ReLU')
# show the plot
pyplot.legend()
pyplot.show()
```

```python
# plot of a sigmoid activation function and its gradient in the same graph
from math import exp
# sigmoid activation function
def sigmoid(x):
    return 1.0 / (1.0 + exp(-x))
# derivative of sigmoid function
def sigmoid_derivative(x):
    return sigmoid(x) * (1.0 - sigmoid(x))

# Define the x range
x_range = np.linspace(-5, 5, 1000)

# line plot of a sigmoid function
pyplot.plot(x_range, [sigmoid(x) for x in x_range], label='sigmoid')
# line plot of derivative of sigmoid function
pyplot.plot(x_range, [sigmoid_derivative(x) for x in x_range], label='Gradient of Sigmoid')
# show the plot
pyplot.legend()
pyplot.show()
```
